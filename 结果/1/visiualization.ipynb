```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import d2l.torch as d2l

d2l.use_svg_display()

data = pd.read_csv('data.csv', header=None)
data = data.values.reshape(-1)
x = {'steps': np.arange(len(data)), 'rewards': data}
df = pd.DataFrame(x)

window_size = 40
df['rewards_moving_avg'] = df['rewards'].rolling(window=window_size).mean()

plt.figure(figsize=(10, 6))

plt.plot(df['steps'].values, df['rewards'].values, color='lightgray', label='Reward')
plt.plot(df['steps'].values, df['rewards_moving_avg'].values, color='blue', label='Moving Average Reward')
plt.title('Reward over Steps')
plt.xlabel('Steps')
plt.ylabel('Reward')
plt.legend()
plt.show()
# åŽ200ä¸ªçš„å¹³å‡
print(np.mean(data[-200:]))
```
